# 04 — Calibration Sensitivity

Portfolio composition and unsafe probability sensitivity sweeps.

```python
import sys
from pathlib import Path

REPO_ROOT = Path.cwd()
if REPO_ROOT.name == "notebooks":
    REPO_ROOT = REPO_ROOT.parent
sys.path.insert(0, str(REPO_ROOT / "src"))

OUT_FIGURES = REPO_ROOT / "outputs" / "figures"
OUT_TABLES  = REPO_ROOT / "outputs" / "tables"
OUT_DATA    = REPO_ROOT / "outputs" / "data"
OUT_LOGS    = REPO_ROOT / "outputs" / "logs"
for d in [OUT_FIGURES, OUT_TABLES, OUT_DATA, OUT_LOGS]:
    d.mkdir(parents=True, exist_ok=True)

import numpy as np, pandas as pd
from run_simulation import SimConfig, load_config
from run_calibration_sensitivity import (
    run_portfolio_sensitivity, run_unsafe_prob_sensitivity, plot_sensitivity
)
cfg = load_config(str(REPO_ROOT / "src" / "params_default.json"))
```

```python
# Portfolio composition sensitivity
port_df = run_portfolio_sensitivity(cfg, OUT_DATA)
print(f"Portfolio sensitivity: {len(port_df)} conditions")
print(f"  Gates unsafe range: [{port_df['Gates_unsafe_rate'].min():.4f}, {port_df['Gates_unsafe_rate'].max():.4f}]")
```

```python
plot_sensitivity(port_df, "label", "Portfolio Composition",
    "Unsafe Deployment Rate by Portfolio Composition", "fig_portfolio_sensitivity", OUT_FIGURES)
print("Portfolio sensitivity figure saved.")
```

```python
# Unsafe probability sensitivity
prob_df = run_unsafe_prob_sensitivity(cfg, OUT_DATA)
print(f"Unsafe prob sensitivity: {len(prob_df)} conditions")
print(f"  Gates unsafe range: [{prob_df['Gates_unsafe_rate'].min():.4f}, {prob_df['Gates_unsafe_rate'].max():.4f}]")
```

```python
plot_sensitivity(prob_df, "label", "Unsafe Probability",
    "Unsafe Deployment Rate by Unsafe Probability", "fig_unsafe_prob_sensitivity", OUT_FIGURES)
print("Unsafe probability sensitivity figure saved.")
```
